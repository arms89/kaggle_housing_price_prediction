```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)

h_data_path = 'train.csv'

dump = pd.read_csv(h_data_path, index_col=['Id'])
data = dump.copy()
data.head()
```

```python
for column in data.columns:
    if data[column].isna().sum() > 0:
        print(column, data[column].isna().sum(), sep=' : ')

# obj_columns = [col for col in data.columns if data[col].dtype == 'object' ]

# for column in obj_columns:
#     if data[column].isna().sum() > 0:
#         data[column].fillna('NA', inplace=True)
        
# non_object_columns = [column for column in data.columns if data[column].dtype != object]

# for column in non_object_columns:
#     if data[column].isna().sum() > 0:
#         data[column].fillna(0, inplace=True)
```

```python
data['GarageType'] = data['GarageType'].fillna('Na')
data['GarageYrBlt'] = data['GarageYrBlt'].fillna(0)
data['GarageFinish'] = data['GarageFinish'].fillna('Na')
data['GarageQual'] = data['GarageQual'].fillna('Na')
data['GarageCond'] = data['GarageCond'].fillna('Na')
data['BsmtQual'] = data['BsmtQual'].fillna('Na')
data['BsmtCond'] = data['BsmtCond'].fillna('Na')
data['BsmtExposure'] = data['BsmtExposure'].fillna('Na')
data['BsmtFinType1'] = data['BsmtFinType1'].fillna('Na')
data['BsmtFinType2'] = data['BsmtFinType2'].fillna('Na')
data['Alley'] = data['Alley'].fillna('Na')
data['FireplaceQu'] = data['FireplaceQu'].fillna('Na')
data['PoolQC'] = data['PoolQC'].fillna('Na')
data['Fence'] = data['Fence'].fillna('Na')
data['MiscFeature'] = data['MiscFeature'].fillna('Na')
data['Electrical'] = data['Electrical'].fillna(data['Electrical'].mode()[0])
data = data[data['MasVnrType'].notna()]
data['MSSubClass'] = data['MSSubClass'].astype(str)
```

```python
data_new= pd.get_dummies(data)
data_new.head()
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer()

columns=data_new.columns
```

```python
data_new = pd.DataFrame(imputer.fit_transform(data_new), columns=columns)
```
